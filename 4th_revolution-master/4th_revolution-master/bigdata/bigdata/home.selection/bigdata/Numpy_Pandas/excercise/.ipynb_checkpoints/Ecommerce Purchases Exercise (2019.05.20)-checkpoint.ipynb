** What are the 5 most common Job Titles? **

```python
ecom.Job.value_counts()[:5]
```

```python
jobs={}
for i in ecom['Job']:
    if i in jobs:
        jobs[i]+=1
    else:
        jobs[i]=1

sorted(jobs.items(),key = lambda x: x[1])[-5:]
```

** Hard: What are the top 5 most popular email providers/hosts (e.g. gmail.com, yahoo.com, etc...) **

```python
dic={}
for i in ecom['Email']:
    if i.split('@')[1] in dic:
        dic[i.split('@')[1]]+=1
    else:
        dic[i.split('@')[1]]=1

sorted(dic.items(),key=lambda x: x[1],reverse=True)[:5]
```

```python
email_provider=ecom['Email'].map(lambda x: x.split('@')[1])
```

```python
ecom['email_provider'] = email_provider
```

```python
ecom['email_provider'].value_counts()[:5]
```
