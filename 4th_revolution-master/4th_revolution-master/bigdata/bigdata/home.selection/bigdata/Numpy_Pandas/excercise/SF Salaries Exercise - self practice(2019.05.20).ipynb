** How many Job Titles were represented by only one person in 2013? (e.g. Job Titles with only one occurence in 2013?) **

```python
sal_2013 = sal[sal['Year']==2013]
```

```python
jobs ={}
for i in sal_2013['JobTitle']:
    if i in jobs:
        jobs[i]+=1
    else:
        jobs[i]=1
print(jobs)

count=0
for i in list(jobs.values()):
    if i == 1:
        count+=1
print(count)
```
