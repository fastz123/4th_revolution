## Pivot

```python
df_phone = pd.read_csv("./phone_data.csv")
```

```python
df_phone['date']=df_phone['date'].apply(dateutil.parser.parse,dayfirst=True)
```

```python
df_phone.head()
```

```python
df_phone.pivot_table(['duration'],index=[df_phone.month,df_phone.item],columns=df_phone.network,aggfunc='sum',fill_value=0)
```

```python
df_phone.pivot_table(['duration'],index=[df_phone.month],columns=df_phone.network,aggfunc='sum',fill_value=0)
```

## crosstab

```python
df_movie = pd.read_csv("data/movie_rating.csv")
df_movie.head()
```

```python
df_movie.pivot_table(["rating"],index=df_movie.critic,columns=df_movie.title,aggfunc='sum',fill_value=0)
```

```python
df_movie.pivot_table(["rating"],index=df_movie.critic,columns=df_movie.title,aggfunc='mean',fill_value=0)
```

```python
pd.crosstab(index=df_movie.critic,columns=df_movie.title,values=df_movie.rating,aggfunc='first').fillna(0)
```

```python
df_movie.groupby(["critic","title"]).agg({"rating":"first"})
```

```python
df_movie.groupby(["critic","title"]).agg({"rating":"first"}).unstack().fillna(0)
```

## merge

```python
raw_data = {
'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_a = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_a
```

```python
raw_data = {
'subject_id': ['4', '5', '6', '7', '8'],
'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_b = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_b
```

```python
pd.merge(df_a,df_b,on='subject_id')
```

```python
pd.merge(df_a,df_b,left_on='subject_id',right_on='subject_id')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='left')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='right')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='outer')
```

```python
pd.merge(df_a, df_b, on='subject_id', how='inner')
```

```python
pd.merge(df_a, df_b, right_index=True, left_index=True)
```
