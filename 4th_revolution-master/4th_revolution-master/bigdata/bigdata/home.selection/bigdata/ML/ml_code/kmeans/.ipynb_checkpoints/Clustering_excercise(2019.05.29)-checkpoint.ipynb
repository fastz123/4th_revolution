```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=150,
                    n_features=2,
                    centers=3,
                    cluster_std=0.5,
                    shuffle=True,
                    random_state=0)
```

```python
print(X,y)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1],
c='white', marker='o', edgecolor='black', s=50)
plt.grid()
plt.tight_layout()
#plt.savefig('images/11_01.png', dpi=300)
plt.show()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3,
            init='random',
            max_iter = 300,
            random_state=0)
y_km = km.fit_predict(X)
```

```python
print(y_km)
```

```python
X[y_km == 0, 0]
```

```python
X[y_km == 1, 0]
```

```python
print(km.cluster_centers_)
```

```python
plt.scatter(X[y_km == 0, 0],
            X[y_km == 0, 1],
            s=50, c='lightgreen',
            marker='s', edgecolor='black',
            label='cluster 1')
plt.scatter(X[y_km == 1, 0],
            X[y_km == 1, 1],
            s=50, c='orange',
            marker='o', edgecolor='black',
            label='cluster 2')
plt.scatter(X[y_km == 2, 0],
            X[y_km == 2, 1],
            s=50, c='lightblue',
            marker='v', edgecolor='black',
            label='cluster 3')
plt.scatter(km.cluster_centers_[:, 0],
            km.cluster_centers_[:, 1],
            s=250, marker='*',
            c='red', edgecolor='black',
            label='centroids')
plt.legend(scatterpoints=1)
plt.grid()
plt.tight_layout()
#plt.savefig('images/11_02.png', dpi=300)
plt.show()
```

```python
print('Distortion: %.2f' % km.inertia_)
```

```python
distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i,
            init='random',
            random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
#plt.savefig('images/11_03.png', dpi=300)
plt.show()
```

```python
from sklearn.cluster import KMeans
km = KMeans(n_clusters=3,
        init='k-means++',
        max_iter = 300,
        random_state=0)
y_km = km.fit_predict(X)
```

# silhouette
* 군집내 샘플들이 얼마나 결합력 있게 그룹핑이 되었는지
* a - 샘플의 군집 내의 모든 점들과의 평균거리, 군집내 응집력(작을수록)
* b - 샘플과 가장 가까운 군집의 모든 샘플들과의 평균거리, 군집 분리력(클수록)

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_samples
```

```python
km = KMeans(n_clusters=3,init='k-means++',random_state=0)
y_km = km.fit_predict(X)
```

```python
print(X[:5])
```

```python
print(y_km[:5])
```

```python
cluster_labels=np.unique(y_km)
```

```python
print(cluster_labels)
```

```python
n_clusters = cluster_labels.shape[0]
```

```python
silhouette_vals = silhouette_samples(X,y_km,metric='euclidean')
```

```python
print(silhouette_vals[:5])
```

```python
print(len(silhouette_vals))
```

```python
y_ax_lower,y_ax_upper=0,0
yticks=[]
```

```python
np.mean(silhouette_samples(X,y_km,metric='euclidean'))
```

```python
plt.plot(range(len(silhouette_vals)),sorted(silhouette_vals))
```

```python
clist=['g','b','y']
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    # color = cm.jet(float(i) / n_clusters)b
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0,
    edgecolor='none', color=clist[i])
    yticks.append((y_ax_lower + y_ax_upper) / 2.)
    y_ax_lower += len(c_silhouette_vals)
    silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color="red", linestyle="--")
plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')
plt.tight_layout()
#plt.savefig('images/11_06.png', dpi=300)
plt.show()

```

```python
km = KMeans(n_clusters=2,
        init='k-means++',
        n_init=10,
        max_iter=300,
        tol=1e-04,
        random_state=0)
y_km = km.fit_predict(X)


plt.scatter(X[y_km == 0, 0],
        X[y_km == 0, 1],
        s=50,
        c='lightgreen',
        edgecolor='black',
        marker='s',
        label='cluster 1')
plt.scatter(X[y_km == 1, 0],
        X[y_km == 1, 1],
        s=50,
        c='orange',
        edgecolor='black',
        marker='o',
        label='cluster 2')

plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1],
s=250, marker='*', c='red')#, label='centroids')
plt.legend()
plt.grid()
plt.tight_layout()
#plt.savefig('images/11_05.png', dpi=300)
plt.show()
```

```python
cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(X, y_km, metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []
colorlist=['g','b']
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    # color = cm.jet(float(i) / n_clusters)b
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0,
    edgecolor='none', color=colorlist[i])
    yticks.append((y_ax_lower + y_ax_upper) / 2.)
    y_ax_lower += len(c_silhouette_vals)
    silhouette_avg = np.mean(silhouette_vals)
plt.axvline(silhouette_avg, color="red", linestyle="--")
plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')
plt.tight_layout()
#plt.savefig('images/11_06.png', dpi=300)
plt.show()
```

```python
#<문제> 최적의 클러스터링 갯수를 silhouette 기반으로 확인하시오
silhouette_list=[]
for i in range(2,11):
    km=KMeans(n_clusters=i,init='k-means++',n_init=10,max_iter=300,tol=1e-04,random_state=0)
    y_km = km.fit_predict(X)
    
    silhouette_vals = silhouette_samples(X,y_km,metric='euclidean')
    silhouette_avg = np.mean(silhouette_vals)
    silhouette_list.append(silhouette_avg)

plt.plot(range(2,11),silhouette_list)
print(np.argmax(silhouette_list)+2)
```

# Hierachical Clustering - Agglomerative Clustering

1. 모든 샘플들의 거리 행렬을 계산
2. 한 계체 군집으로 개별 관측치를 표시한다
3. 가장 멀리있는 개체들의 거리에 기반하여 가장 가까운 2개의 군집을 합친다.
4. 유사도 행렬을 업데이트한다.
5. 하나의 군집만 남을 때까지 반복한다.
6. ward - 모든 클러스터 내의 분산을 가장 작게 증가시키는 두 클러스터를 합친다.
7. average - 클러스터 사이의 평균 거리가 가장 짧은 두 클러스터를 합친다.
8. complete - 클러스터 포인트 사이의 최대거리가 가장 짧은 두 클러스터를 합친다.

```python
import mglearn
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
```

```python
X,y = make_blobs(random_state=1)
agg = AgglomerativeClustering(n_clusters=3)
assignment = agg.fit_predict(X)
print(assignment[:20])

mglearn.discrete_scatter(X[:,0],X[:,1],assignment)
plt.legend(['0','1','2'],loc='best')
plt.show()
```

```python
mglearn.plots.plot_agglomerative()
```

```python
from scipy.cluster.hierarchy import dendrogram, ward
linkage_array = ward(X) #
row_dendr = dendrogram(linkage_array, # dendrogram
labels=y,
)
plt.tight_layout()
plt.ylabel('Euclidean distance')
plt.show()
```

# Locating regions of high density via DBSCAN
## 밀도기반 클러스터링

* 밀도 - 특정반지름 e 내에 점들의 수
* 설정한 이웃 점의 개수가 설정한 반지름 내에 존재한다면 그점은 중심점
* 경게점 - 반지름 내에서 설정한 이웃 점의 개수 보다 더 작은 수의 이웃을 갖는 점이지만 중심점으로 부터 반지름 내에 위치
* 잡음점 - 중심점도 경계점도 아닌 다른 모든 점들
* 중심점이나 중심점들이 연결된 그룹에 대해 군집을 형성, 반지름보다 떨어져 있지 않다면 중심점들은 연결된다.
* 각각의 경계점을 자신의 중심점이 포함되는 군집에 할당
* 장점 - 군집의 모양에 대한 가정이 불필요. 각각의 점을 군집에 할당할 필요 없음, 잡음의 제거가 쉽다.
* 단점 - 차원이 증가하면 부정정 영향
* 초기에 random한 값으로 시작, 밀집 지역 합쳐나감, 중복되는 경계점은 먼저 방문한 중심점에 할당

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score

db = DBSCAN(eps=0.2,min_samples=5,metric='euclidean')
y_db = db.fit_predict(X)
#print(y_db)
plt.scatter(X[y_db==0, 0],X[y_db ==0, 1], c='lightblue',marker='o',s=40,edgecolor='black',label='cluster 1')
plt.scatter(X[y_db==1, 0],X[y_db ==1, 1], c='red',marker='s',s=40,edgecolor='black',label='cluster 2')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
#<문제> kmeans와 agg 로 silhouette 검증
# kmeans

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

km=KMeans(n_clusters=2,init='k-means++',n_init=10,max_iter=300,tol=1e-04,random_state=0)
y_km = km.fit_predict(X)
#print(y_db)
plt.scatter(X[y_km==0, 0],X[y_km ==0, 1], c='lightblue',marker='o',s=40,edgecolor='black',label='cluster 1')
plt.scatter(X[y_km==1, 0],X[y_km ==1, 1], c='red',marker='s',s=40,edgecolor='black',label='cluster 2')
plt.legend()
plt.tight_layout()
plt.show()

sil = silhouette_score()
```

```python
# agglomertive

from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score

agg = AgglomerativeClustering(n_clusters=2)
y_agg = agg.fit_predict(X)
#print(y_db)
plt.scatter(X[y_agg==0, 0],X[y_agg ==0, 1], c='lightblue',marker='o',s=40,edgecolor='black',label='cluster 1')
plt.scatter(X[y_agg==1, 0],X[y_agg ==1, 1], c='red',marker='s',s=40,edgecolor='black',label='cluster 2')
plt.legend()
plt.tight_layout()
plt.show()
```
