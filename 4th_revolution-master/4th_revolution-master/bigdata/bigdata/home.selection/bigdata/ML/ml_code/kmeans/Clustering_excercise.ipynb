# Grouping objects by similarity using k-means

## K-means clustering using scikit-learn

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(n_samples=150, 
                  n_features=2, 
                  centers=3, 
                  cluster_std=0.9, 
                  shuffle=True, 
                  random_state=0)
```

```python
print(X,y)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], 
            c='white', marker='o', edgecolor='black', s=50)
plt.grid()
plt.tight_layout()
#plt.savefig('images/11_01.png', dpi=300)
plt.show()
```

```python
from sklearn.cluster import KMeans

km = KMeans(n_clusters=3, 
            init='random', 
            max_iter = 300,
            random_state=0)
# km.fit(X)
# y_km = km.predict(X)

y_km = km.fit_predict(X)
```

```python
print('Distortion: %.2f' % km.inertia_)
```

```python
# kmeans++ - 초기의 센트로이드를 가능한 멀리 두는 것, 
# 고전적 kmeans 보다 나은 성과
# inertia_ : 군집내 SSE

distortions = []
for i in range(1, 11):
    km = KMeans(n_clusters=i, 
                init='random', 
                random_state=0)
    km.fit(X)
    distortions.append(km.inertia_)
plt.plot(range(1, 11), distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.tight_layout()
#plt.savefig('images/11_03.png', dpi=300)
plt.show()
```

## silhouette
* 군집 내 샘플들이 얼마나 결합력 있게 그룹핑이 되었는지
* a - 샘플의 군집 내의 모든 점들과의 평균거리, 군집내 응집력
* b - 샘플과 가장 가까운 군집의 모든 샘플들과의 평균거리, 군집 분리력
* silhouette = (b-a)/max(b,a)

# Hierachical Clustering - Agglomerative Clustering 
1. 모든 샘플들의 거리 행렬을 계산
2. 한 계체 군집으로 개별 관측치를 표시한다
3. 가장 멀리있는 개체들의 거리에 기반하여 가장 가까운 두 개의 군집을 합친다
4. 유사도 행렬을 업데이트한다.
5. 하나의 군집만 남을 때까지 반복한다

* ward - 모든 클러스터 내의 분산을 가장 작게 증가시키는 두 클러스터를 합칩니다.
* average - 클러스터 사이의 평균 거리가 가장 짧은 두 클러스터를 합칩니다.
* complete - 클러스터 포인트 사이의 최대 거리가 가장 짧은 두 클러스터를 합칩니다.

# Locating regions of high density via DBSCAN
## 밀도기반 클러스터링
* 밀도 - 특정반지름 e 내에 점들의 수
* 설정한 이웃 점의 개수(MinPts)가 설정한 반지름 e 내에 떨어진다면 그점은 중심점
* 경계점 - e 내에서 설정한 이웃 점의 개수(MinPts) 보다 더 작은 수의 이웃을 갖는 점이지만, 중심점으로 부터 반지름 e 내에 위치
* 잡음점 - 중심점도 경계점도 아닌 다른 모든 점들.
* 중심점이나 중심점들이 연결된 그룹에 대해 군집을 형성, e보다 떨어져 있지 않다면 중심점들은 연결된다 (밀집지역)
* 각각의 경계점을 자신의 중심점이 포함되는 군집에 할당
* 장점 - 군집의 모양에 대한 가정이 불필요, 각각의 점을 군집에 할당할 필요 없음, 잡음의 제거가 쉽다
* 단점 - 차원(피쳐수)이 증가하면 부정적 영향 
* 초기에 random한 값으로 시작, 밀집지역 합쳐나감, 중복되는 경계점은 먼저 방문한 중심점에 할당


```python
from sklearn.datasets import make_moons

X, y = make_moons(n_samples=200, noise=0.05, random_state=0)
plt.scatter(X[:, 0], X[:, 1])
plt.tight_layout()
#plt.savefig('images/11_14.png', dpi=300)
plt.show()
```

```python
from sklearn.cluster import DBSCAN
# from sklearn.metrics.cluster import silhouette_score

db = DBSCAN(eps=0.2, min_samples=5, metric='euclidean')
y_db = db.fit_predict(X)

plt.scatter(X[y_db == 0][:, 0], X[y_db == 0][:, 1],
            c='lightblue', marker='o', s=40,
            edgecolor='black', label='cluster 1')

plt.scatter(X[y_db == 1][:, 0], X[y_db == 1][:, 1],
            c='red', marker='s', s=40,
            edgecolor='black', label='cluster 2')

plt.legend()
plt.tight_layout()
plt.show()
```

K-means and hierarchical clustering:

```python
# from sklearn.metrics.cluster import silhouette_score
# f, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 3))

km = KMeans(n_clusters=2, random_state=0)
y_km = km.fit_predict(X)
# print(silhouette_score(X, y_km))

plt.scatter(X[y_km == 0, 0], X[y_km == 0, 1],
            edgecolor='black',
            c='lightblue', marker='o', s=40, label='cluster 1')
plt.scatter(X[y_km == 1, 0], X[y_km == 1, 1],
            edgecolor='black',
            c='red', marker='s', s=40, label='cluster 2')
plt.title('K-means clustering')
```

```python
# from sklearn.metrics.cluster import silhouette_score
ac = AgglomerativeClustering(n_clusters=2,
                             affinity='euclidean',
                             linkage='ward')
y_ac = ac.fit_predict(X)
# print(silhouette_score(X, y_ac))

plt.scatter(X[y_ac == 0, 0], X[y_ac == 0, 1], c='lightblue',
            edgecolor='black',
            marker='o', s=40, label='cluster 1')
plt.scatter(X[y_ac == 1, 0], X[y_ac == 1, 1], c='red',
            edgecolor='black',
            marker='s', s=40, label='cluster 2')
plt.title('Agglomerative clustering')

plt.legend()
plt.tight_layout()
# plt.savefig('images/11_15.png', dpi=300)
plt.show()
```

Density-based clustering:

```python
from sklearn.metrics.cluster import silhouette_score
from sklearn.datasets import make_moons
from sklearn.preprocessing import StandardScaler
import numpy as np
import mglearn
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import DBSCAN




X, y = make_moons(n_samples=200, noise=0.05, random_state=0)

# 평균이 0, 분산이 1이 되도록 데이터의 스케일을 조정합니다
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)

fig, axes = plt.subplots(1, 4, figsize=(15, 3),
                         subplot_kw={'xticks': (), 'yticks': ()})

# 비교를 위해 무작위로 클러스터 할당을 합니다
random_state = np.random.RandomState(seed=0)
random_clusters = random_state.randint(low=0, high=2, size=len(X))

# 무작위 할당한 클러스터를 그립니다
axes[0].scatter(X_scaled[:, 0], X_scaled[:, 1], c=random_clusters,
                cmap=mglearn.cm3, s=60, edgecolors='black')
axes[0].set_title("무작위 할당: {:.2f}".format(
        silhouette_score(X_scaled, random_clusters)))

algorithms = [KMeans(n_clusters=2), AgglomerativeClustering(n_clusters=2),
              DBSCAN()]

for ax, algorithm in zip(axes[1:], algorithms):
    clusters = algorithm.fit_predict(X_scaled)
    # 클러스터 할당과 클러스터 중심을 그립니다
    ax.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap=mglearn.cm3,
               s=60, edgecolors='black')
    ax.set_title("{} : {:.2f}".format(algorithm.__class__.__name__,
                                      silhouette_score(X_scaled, clusters)))
```
