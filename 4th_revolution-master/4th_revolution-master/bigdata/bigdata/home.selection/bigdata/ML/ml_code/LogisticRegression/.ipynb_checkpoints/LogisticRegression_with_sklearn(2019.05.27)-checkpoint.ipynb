```python
import pandas as pd
```

```python
data_url='uva.txt'
```

```python
dataframe = pd.read_table(data_url)
```

```python
dataframe.pop('who')
```

```python
dataframe.pop('Country')
dataframe.pop('Years on Internet')
```

```python
dataframe['Gender']=dataframe['Gender'].astype('category')
dataframe.dtypes
```

```python
for col in ['Household Income','Sexual Preference','Education Attainment','Major Occupation','Marital Status']:
    dataframe[col] = dataframe[col].astype('category')
dataframe.dtypes
```

```python
dataframe.head()
```

```python
dataframe_modified = pd.get_dummies(dataframe)
dataframe_modified[:5]
```

```python
dataframe_modified.shape
```

```python
dataframe_modified.isnull().sum()
```

```python
dataframe_modified.head()
```

```python
dataframe_modified['Age']=dataframe_modified['Age'].fillna(dataframe_modified['Age'].mean())
```

```python
dataframe_modified.info()
```

```python
dataframe_modified['Age'].isnull().sum()
```

```python
dataframe_modified['Age']=dataframe_modified['Age'].astype('int')
```

```python
dataframe_modified.head()
```

```python
x = dataframe_modified.iloc[:,1:]
y=dataframe_modified.iloc[:,0]
```

```python
from sklearn import preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
x_data = min_max_scaler.fit_transform(x)
x[:5]
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_data,y,test_size=0.2)
```

```python
from sklearn import linear_model
logmodel = linear_model.LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
x_train.shape,y_train.shape
```

```python
predictions = logmodel.predict(x_test)
```

```python
logmodel.predict_proba(x_test[:10])
```
